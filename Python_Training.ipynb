```python
##merging and joining DataFrames in python
```

```python
"""creating a new data frame
"""

xdata = pd.DataFrame({
'Country': ['Kenya', 'Uganda', 'Tanzania'],
    'GDP': [21, 1.7, 4.2]
    
})
xdata
```

```python
"""merging is done based on common column, in this case, its "Country"
"""
""" xdata DataFrame and data DataFrame are merged in this case
"""
merged_df = pd.merge(
    data,xdata, on = 
    'Country', how =  'inner'
)
merged_df
```

```python
"""
Joining works the same but with index
the common column has to be set to an index first
using the .set_index() method
df1.set_index("Column").join(df2.set_index("key"))
"""
data.set_index("Country").join(
    xdata.set_index("Country"), how = 'left').head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
car1 = pd.read_excel('car_data.xlsx')
car1
```

```python
##creating multiple scatter plots
```

```python
##combining seaborn, matplotlib pandas to create multiple scatter plots
```

```python
variables = ['Min Price (Lakh)', 'Max Price (Lakh)', 	'Range (kmpl)', 'CC',	'Seats',	'Variants',	'Ex-Showroom Price',	'RTO',	'Insurance',	'Other','Onroad Price']
```

```python
plt.figure(figsize= (18, 15))
for i, var in enumerate(variables, start = 1):
    plt.subplot(4, 3,i)
    sns.scatterplot(x = car1[var], y = car1['Max Price (Lakh)'], alpha = 0.7)
    plt.title(f"scatter plot of {var} vs Onroad Price")
    plt.xlabel(var)
    plt.ylabel('Target')
    plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20, 15))
for i, var in enumerate(variables, start=1):
    plt.subplot(4, 3, i)
    
    # Using sns.regplot instead of sns.scatterplot to include a line of best fit
    sns.regplot(x=car1[var], y=car1['Max Price (Lakh)'], scatter_kws={'alpha':0.7})
    
    plt.title(f"Scatter plot of {var} vs Onroad Price")
    plt.xlabel(var)
    plt.ylabel('Target')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
##multiple histograms
```

```python
plt.figure(figsize= (18, 15))
for i, var in enumerate(variables, start = 1):
    
    plt.subplot(4, 3,i)
    sns.histplot(car1[var], bins = 20, kde = True, color = "Green")
    plt.title(f"histogram of {var}")
    plt.xlabel(var)
    plt.ylabel('Frequency')
    plt.xticks(rotation = 45)
    
plt.tight_layout()
plt.show()
```

#Heat Maps
"""
Great for visualizing correlation matrices.
You can use them to undrstand the relation between
different variables.
"""

```python
selected_columns = ['Min Price (Lakh)','Max Price (Lakh)','Range (kmpl)',
                    'CC', 'Seats', 'Variants', 
                    'Ex-Showroom Price','RTO',	'Insurance','Other']
df_selected = car1[selected_columns]
df_selected
```

```python
#calculating the correlation btwn the variables
corr = df_selected.corr()
corr
```

"""
Plotting heat maps to visualize the correaltions
"""

```python
sns.heatmap(corr, annot = True,cmap = 'Greens')
plt.title("Heat map of Correaltion Matrix")
plt.show()
```


The colour green increases intensity as the 
correlation edges towards +1 and decreases as the corr edges towards -1
The annot is added to include the numbers in the map.


```python
sns.heatmap(corr,cmap = 'Greens')
plt.title("Heat map of Correaltion Matrix")
plt.show()
```
